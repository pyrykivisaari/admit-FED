# Example: Local densities of states

Here, we calculate the optical local densities of states (LDOS) in the simplified InGaN/GaN multi-quantum well (MQW) light-emitting diode (LED) structure shown below.

The LDOS functions shown in the figure below are the ones corresponding to Eqs. (S5) and (S6) in the supplementary material of https://doi.org/10.1038/s41598-017-11753-5 (Open access). The dashed vertical lines in the figures mark certain interfaces of interest: the ones on the left-hand side mark the edges of the MQW region, and the one on the right-hand side the location of the thin silver contact that separates GaN from air.

Plotted like this with arbitrary units, the LDOS can essentially be used to visually compare the availability of optical modes at different segments in the figures. For example, with in-plane $K$ numbers between 0 and $k_0$, propagating modes exist in all the layers spanned by the $x$ axis. With $K$ between $k_0$ and $2.5k_0$, propagating modes exist only in GaN and InGaN, whereas the LDOS in air only showcases the availability of evanescent modes, which diminishes when moving away from the Air/GaN interface. With $K$ larger than $k_0$, the LDOS only includes fully evanescent modes even in GaN and InGaN.

# Example: reflection coefficient calculation

Here, we calculate the reflectivity of a simplified contact structure of a GaAs/AlGaAs optoelectronic device (such as a solar cell or LED). This is a very standard problem in optics most often dealt with by using the transfer matrix method. However, the optical admittances allow calculating the reflection coefficients as a function of position everywhere in the geometry, not only at the interfaces.

The example structure studied here is shown below. For simplicity, it assumes light to enter from a lossless material with the real refractive index close to that of GaAs. Then, light enters a 400 nm thick cladding layer of GaAs before hitting a thin GaAs contact layer and the Ag contact metal.

The power reflectances are shown in the figures further below for a photon energy 1.43 eV, which is close to the GaAs band gap energy. The reflectances are plotted as a function of propagation angle (in the lossless GaAs) separately for TE and TM modes. The curves exhibit some resonant effects due to the AlGaAs layer. The thickness of the GaAs contact layer has a big influence in the reflectivity due to its absorptivity (the Ag contact also absorbs part of the light). The refractive index of AlGaAs causes the total internal reflection angle between GaAs and AlGaAs to be around 70 degrees, making the reflectance to go to 1 at that angle.

# Example: emission of a semiconductor layer

Here, we finally deploy the Green's functions to something they're actually made for: calculating the emission power of a slab of material inside a specific optical environment. To start with a simple example, the image below shows a 300 nm thick slab of GaAs inside semi-infinite layers of AlGaAs. In the calculation, we assume that the GaAs layer has a temperature of 300 K and is biased at 1.3 V (for readers more specialized in semiconductor optics: this means that the quasi-Fermi level separation is set to 1.3 V throughout the GaAs layer).

The images further below visualize the emission as a function of both photon energy and propagation angle. The first one showcases the total emission integrated over the GaAs layer, and the second one illustrates emission towards the left. As one might expect in the case of this symmetric structure, emission towards the left is half of the total emission power. The last image shows the emission spectrum calculated by integrating over all directions. By integrating the spectrum over photon energy, one gets the total emission power density given in the end of the text output.

More explanations and related results can be found in https://arxiv.org/abs/2401.05108 (Open-access preprint of the paper published in the Proceedings of SPIE).
